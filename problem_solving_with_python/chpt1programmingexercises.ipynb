```python
def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n

        m = oldn
        n = oldm%oldn
    return n

class Fraction:
    def __init__(self,top,bottom):
        if type(top) != int or type(bottom) != int:
            raise ValueError
            print "Numerator and Denominator must be Integers"
        if top < 0 and bottom < 0:
            top = top*-1
            bottom = bottom*-1
        elif bottom < 0 and top >= 0:
            top = top*-1
            bottom = bottom*-1
            
        common = gcd(top,bottom)
        self.num = top//common
        self.den = bottom//common
            
    def __str__(self):
        return str(self.num)+"/"+str(self.den)
    
    def __repr__(self):
        return "Fraction Object Numrator={num} Denominator={den}".format(num=self.num,den=self.den)

    def show(self):
        print(self.num,"/",self.den)

    def __add__(self,otherfraction):
        if type(otherfraction) == int:
            otherfraction = Fraction(otherfraction*self.den,self.den)
        try:
            newnum = self.num*otherfraction.den + \
            self.den*otherfraction.num
            newden = self.den * otherfraction.den
            return Fraction(newnum,newden)
        except TypeError:
            print "Can only add ints and other Fraction objects for now"
    
    def __radd__(self,other):
        if type(other) == int:
            otherfraction = Fraction(other*self.den,self.den)
            return Fraction(self.num,self.den) + otherfraction
        else:
            raise TypeError
            print "Can only add ints to fractions....for now"
    
    def __sub__(self,otherfraction):
        newnum = (self.num*otherfraction.den - otherfraction.num*self.den)
        newden = self.den*otherfraction.den
        return Fraction(newnum,newden)

    def __eq__(self, other):
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum == secondnum
    
    def __ne__(self, other):
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return not firstnum == secondnum
    
    def __gt__(self, other):
        return self.num*other.den > self.den*other.num
    
    def __ge__(self, other):
        return self.num*other.den >= self.den*other.num
    
    def __lt__(self, other):
        return self.num*other.den < self.den*other.num
    
    def __le__(self, other):
        return self.num*other.den <= self.den*other.num
       
    def getNum(self):
        return self.num
    
    def getDen(self):
        return self.den
    
    
x = Fraction(5,10)
y = Fraction(32,48)
z = Fraction(1,2)
print(x)
print(y)
print(x+y)
print(x-y)
print(y-x)
print(x-z)
print(x == z)
print(x != z)
print(x != y)
print(x>y)
print(x<y)
print(y>x)
print(y<x)
print(x>z)
print(x>=z)
print(x<z)
print(x<=x)
Fraction(11,12)
print(x+4)
```

```python
class LogicGate(object):
    """Base Class for a gate"""
    
    def __init__(self,n):
        object.__init__(self)
        self.name = n
    
    @property
    def getName(self):
        return self.name
    
    def getOutput(self):        
        self.output = self.performGateLogic()        
        return self.output
    
    def __str__(self):
        return self.name
    

    
class BinaryGate(LogicGate):
    
    def __init__(self,n):
        LogicGate.__init__(self,n)
        self.pinA = None
        self.pinB = None

        
    def getInput(self,a,b):
        self.pinA = a
        self.pinB = b
    
        
class AndGate(BinaryGate):
    
    def __init__(self,n):
        BinaryGate.__init__(self, n)

    
    def performGateLogic(self):
        
        if self.pinA == None:
            self.pinA = input('Enter input for pin A')
        
        if self.pinB == None:
            self.pinB = input('Enter input for pin B')
        
        if self.pinA == 1 and self.pinB == 1:
            return 1
        else:
            return 0

    def __str__(self):
        return "AndGate {name} with inputs {pinA}:{pinB}".format(name=self.name,
                                                              pinA=self.pinA,pinB=self.pinB)
        

class NAndGate(BinaryGate):
    
    def __init__(self,n):
        BinaryGate.__init__(self, n)
    
    def performGateLogic(self):
        
        if self.pinA == None:
            self.pinA = input('Enter input for pin A')
        
        if self.pinB == None:
            self.pinB = input('Enter input for pin B')
        
        if self.pinA == 1 and self.pinB == 1:
            return 0
        else:
            return 1

    def __str__(self):
        return "NAndGate {name} with inputs {pinA}:{pinB}".format(name=self.name,
                                                              pinA=self.pinA,pinB=self.pinB)


class OrGate(BinaryGate):
    
    def __init__(self,n):
        BinaryGate.__init__(self, n)
    
    def performGateLogic(self):
        
        if self.pinA == None:
            self.pinA = input('Enter input for pin A')
        
        if self.pinB == None:
            self.pinB = input('Enter input for pin B')
        
        if self.pinA == 1 or self.pinB == 1:
            return 1
        else:
            return 0

    def __str__(self):
        return "OrGate {name} with inputs {pinA}:{pinB}".format(name=self.name,
                                                              pinA=self.pinA,pinB=self.pinB)

        
class NOrGate(BinaryGate):
    
    def __init__(self,n):
        BinaryGate.__init__(self, n)
    
    def performGateLogic(self):
        
        if self.pinA == 1 or self.pinB == 1:
            return 0
        else:
            return 1

    def __str__(self):
        return "OrGate {name} with inputs {pinA}:{pinB}".format(name=self.name,
                                                              pinA=self.pinA,pinB=self.pinB)

class XOrGate(BinaryGate):
    
    def __init__(self,n):
        BinaryGate.__init__(self, n)
    
    def performGateLogic(self):

        if self.pinA == None:
            self.pinA = input('Enter input for pin A')
        
        if self.pinB == None:
            self.pinB = input('Enter input for pin B')
        
        if (self.pinA == 1 and self.pinB == 0) or (self.pinA == 0 and self.pinB == 1):
            return 1
        else:
            return 0


class UnaryGate(LogicGate):
    
    def __init__(self, n):
        LogicGate.__init__(self, n)
        self.pin = None
    
    def getInput(self,a):
        self.pin = a
    

class NotGate(UnaryGate):
    
    def __init__(self, n):
        UnaryGate.__init__(self, n)
    
    def performGateLogic(self):
        
        if self.pin == None:
            self.pin = input('Enter input for this gate')
        
        if self.pin == 1:
            return 0
        else:
            return 1

    def __str__(self):
        return "NotGate {name} with input {pin}".format(name=self.name,pin=self.pin)

class Connector:
    
    def __init__(self,topin):
        
        self.topin = topin
    

class BinaryConnector(Connector):
    
    def __init__(self,topin,firstinput,secondinput):
        Connector.__init__(self,topin)
        
        self.topin.getInput(firstinput.getOutput(),secondinput.getOutput())
        
class UnaryConnector(Connector):
    
    def __init__(self,topin,firstinput):
        Connector.__init__(self, topin)
        
        self.topin.getInput(firstinput.getOutput())


def halfadder(a,b):
    carry = AndGate('Carry')
    carry.getInput(a,b)
    output = XOrGate('Output')
    output.getInput(a,b)
    return carry.getOutput(), output.getOutput()

def fulladder(a,b,cin):
    carry1, sum1 = halfadder(a,b)
    carry2, sum2 = halfadder(sum1,cin)
    finalcarry = OrGate('FinalCarry')
    finalcarry.getInput(carry2,carry1)
    
    return finalcarry.getOutput(), sum2


print(fulladder(0,0,0))
print(fulladder(0,0,1))
print(fulladder(0,1,0))
print(fulladder(0,1,1))
print(fulladder(1,0,0))
print(fulladder(1,0,1))
print(fulladder(1,1,0))
print(fulladder(1,1,1))

        
```

```python
class Card:
    def __init__(self,suit,number):
        self.suit = suit
        self.number = number

    @property
    def suit(self):
        return self.suit
    
    @property
    def number(self):
        return self.number
    
    def __str__(self):
        return "{number} of {suit}".format(number=self.number,suit=self.suit)

class Deck:
    
    def __init__(self):
        suits = ['hearts','spades','clovers','diamons']
        numbers = range(1,14)
    
        self.deck = [Card(suit,number)
                for suit in suits
                for number in numbers]
        
    def show_deck(self):
        for card in self.deck:
            print(card)
    
    def get_deck(self):
        return self.deck
    

def main():
    mydeck = Deck()
    mydeck.show_deck()

main()
```
