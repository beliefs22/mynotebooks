```python
def gcd(m,n):
    while m%n != 0:
        oldm = m
        oldn = n

        m = oldn
        n = oldm%oldn
    return n

class Fraction:
    def __init__(self,top,bottom):
        if type(top) != int or type(bottom) != int:
            raise ValueError
            print "Numerator and Denominator must be Integers"
        if top < 0 and bottom < 0:
            top = top*-1
            bottom = bottom*-1
        elif bottom < 0 and top >= 0:
            top = top*-1
            bottom = bottom*-1
            
        common = gcd(top,bottom)
        self.num = top//common
        self.den = bottom//common
            
    def __str__(self):
        return str(self.num)+"/"+str(self.den)
    
    def __repr__(self):
        return "Fraction Object Numrator={num} Denominator={den}".format(num=self.num,den=self.den)

    def show(self):
        print(self.num,"/",self.den)

    def __add__(self,otherfraction):
        if type(otherfraction) == int:
            otherfraction = Fraction(otherfraction*self.den,self.den)
        try:
            newnum = self.num*otherfraction.den + \
            self.den*otherfraction.num
            newden = self.den * otherfraction.den
            return Fraction(newnum,newden)
        except TypeError:
            print "Can only add ints and other Fraction objects for now"
    
    def __radd__(self,other):
        if type(other) == int:
            otherfraction = Fraction(other*self.den,self.den)
            return Fraction(self.num,self.den) + otherfraction
        else:
            raise TypeError
            print "Can only add ints to fractions....for now"
    
    def __sub__(self,otherfraction):
        newnum = (self.num*otherfraction.den - otherfraction.num*self.den)
        newden = self.den*otherfraction.den
        return Fraction(newnum,newden)

    def __eq__(self, other):
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return firstnum == secondnum
    
    def __ne__(self, other):
        firstnum = self.num * other.den
        secondnum = other.num * self.den
        return not firstnum == secondnum
    
    def __gt__(self, other):
        return self.num*other.den > self.den*other.num
    
    def __ge__(self, other):
        return self.num*other.den >= self.den*other.num
    
    def __lt__(self, other):
        return self.num*other.den < self.den*other.num
    
    def __le__(self, other):
        return self.num*other.den <= self.den*other.num
       
    def getNum(self):
        return self.num
    
    def getDen(self):
        return self.den
    
    
x = Fraction(5,10)
y = Fraction(32,48)
z = Fraction(1,2)
print(x)
print(y)
print(x+y)
print(x-y)
print(y-x)
print(x-z)
print(x == z)
print(x != z)
print(x != y)
print(x>y)
print(x<y)
print(y>x)
print(y<x)
print(x>z)
print(x>=z)
print(x<z)
print(x<=x)
Fraction(11,12)
print(x+4)
```

```python
class LogicGate:
    """Represnts base of a gate
    args:
        n (str) = name of the gate
        a (int) = input 1
        b (int) = input 2
    """
    def __init__(self,n,a,b):
        self.name = n
        self.pinA = a
        self.pinB = b

    @property
    def getName(self):
        return self.name

    def getInput(a,b):
        self.pinA = a
        self.pinB = b

    def getOutput(self):
        self.output = self.performGateLogic()
        return self.output


class BinaryGate(LogicGate):
    """A gate that takes two inputs"""

    def __init__(self,n,a,b):
        LogicGate.__init__(self,n,a,b)

    def setNextPin(self,source):
        """Gives the output of this gate for the input of another gate"""
      
        return source.getInput(self.)



class AndGate(BinaryGate):
    

    def __init__(self,n,a,b):
        BinaryGate.__init__(self,n,a,b)

    def performGateLogic(self):

        if self.pinA==1 and self.pinB==1:
            return 1
        else:
            return 0

class NAndGate(BinaryGate):

    def __init__(self,n,a,b):
        BinaryGate.__init__(self,n,a,b)
    
    def performGateLogic(self):
        
        if self.pinA==1 and self.pinB==1:
            return 0
        else:
            return 1

class OrGate(BinaryGate):

    def __init__(self,n,a,b):
        BinaryGate.__init__(self,n,a,b)

    def performGateLogic(self):

        if a ==1 or b==1:
            return 1
        else:
            return 0

class XOrGate(BinaryGate):
    
    def __init__(self,n,a,b):
        BinaryGate.__init__(self,n,a,b)
    
    def performGateLogic(self):
        
        if (self.pinA == 1 and self.pinB==0) or (self.pinA==0 and self.pinB==1):
            return 1
        else:
            return 0

class UnaryGate(LogicGate):

    def __init__(self,n,a):
        LogicGate.__init__(self,a)



    def getPin(self):
        self.pin = a


    def setNextPin(self,source):
        
        return source(self.getOutput())


class NotGate(UnaryGate):

    def __init__(self,n,a):
        UnaryGate.__init__(self,n,a)

    def performGateLogic(self):
        if self.pin == 1:
            return 0
        else:
            return 1


class Connector:

    def __init__(self, fgate, tgate):
        self.fromgate = fgate
        self.togate = tgate

        tgate.setNextPin(fgate)

    def getFrom(self):
        return self.fromgate

    def getTo(self):
        return self.togate

def halfAdder(a,b):
    Carry = AndGate("Carry")
    Sum = XOrGate("Sum")
    return Sum.getOutput(), Carry.getOutput()

def fullAdder(a,b,c):
    #connect a and b to the input of a half adder
    
    Sum, Output1 = halfadder(a,b)
    Sum, Output2 = halfadder(Sum, Cout)
    #and or the two outputs
    final = OrGate(Output1, Output2)
    print(OrGate.getOutput())
    
    

def main():
    g1 = AndGate("G1",1,1)
    g2 = AndGate("G2",1,0)
    g3 = NAndGate("G3",1,0)
    c1 = Connector(g1,g2)
    c2 = Connector(g2,g3)
    print(g3.getOutput())
    halfAdder()
main()
```

```python
class Card:
    def __init__(self,suit,number):
        self.suit = suit
        self.number = number

    @property
    def suit(self):
        return self.suit
    
    @property
    def number(self):
        return self.number
    
    def __str__(self):
        return "{number} of {suit}".format(number=self.number,suit=self.suit)

class Deck:
    
    def __init__(self):
        suits = ['hearts','spades','clovers','diamons']
        numbers = range(1,14)
    
        self.deck = [Card(suit,number)
                for suit in suits
                for number in numbers]
        
    def show_deck(self):
        for card in self.deck:
            print(card)
    
    def get_deck(self):
        return self.deck
    

def main():
    mydeck = Deck()
    mydeck.show_deck()

main()
```
